```python
import json

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import datetime

good_files = 0

i = 0

for k in range(1,7):
    
    filename = str(k)+".json"
    
    with open(filename) as f:
        
        tmp = json.load(f)
        
    df = tmp['datapoints']

    average = []
    
    time = []
    
    for i in range(len(df)):
        
        average.append(df[i]['average'])
        
        date = datetime.datetime.fromtimestamp(df[i]['timestamp']/1000.0)
        
        date = date.strftime('%Y-%m-%d %H:%M:%S')
        
        time.append(date)
        
    if sum(average)/len(average) == 0:
        continue
        
    table = pd.DataFrame({'average': average, 'timestamp': time})
    
    print(table)
    
    print("maximim is",table['average'].max(),"in", time[table['average'].idxmax()])
    
    print("minimum is",table['average'].min(),"in", time[table['average'].idxmin()])
    
    small = table.nsmallest(20, ['average'])
    
    large = table.nlargest(20, ['average'])
    
    print(small)
    
    print(large)
    
    plt.figure(figsize=(20,10))

    t = np.arange(0., 10085, 5)
    
    plt.plot(t,average,'r--')
    
    plt.xlabel('timestamp')
    
    plt.ylabel('average')
    
    plt.title('Load on cluster')
    
    good_files += 1
    
print("Amount of files:", 6)

print("Amount of good files:", good_files)
```
